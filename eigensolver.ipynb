# Eigenvalues and Eigenvectore

Eigenvalue &  eigenvector are probably one of the most important concepts in linear algebra. Who can expect a simple equation like Av = λv is so significant? From machine learning, quantum computing, and physic, many mathematical and engineering problems can be solved by finding the eigenvalue and eigenvectors of a matrix. Let’s not only discover what it is but also visualize it using Python.

First let us import necessary packages of python
**Numpy**: for mathematics operations.
**pandas**: for mathematics and data structures.
**matplotlib**: for drawing plots and reframing
**pylab**: PyLab is a convenience module that bulk imports matplotlib.pyplot (for plotting) and NumPy (for Mathematics            and working with arrays) in a single name space.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams #matplotlib inline
rcParams['figure.figsize'] = 8,8
```

for example we have a matrix showing point 1 (x1, y1) and point 2 (x2, y2) \begin{matrix} 9 & 4 \\ 4 & 3 \end{matrix}

```python
# creating a 2 dimentional space in x and y axis from -1 to 1 with 100 divisions

x = np.linspace(-1,1,100)
#plt.draw()
```

writing any equation i.e. equation of a unit circle with x and y coordinates
whose eigenvalues and eigenvectors we are going to find.

x1^2 + y1^2 = 1 or sq(x1) + sq(y1) = 1

=> y1 = squareroot(1- x^2)

y2 is the negative of y1 results we will get from the previous equation

```python
y1 = np.sqrt(1-np.square(x))
y2 = -1 * y1
```

 Here we try to visualize our functoin (circle equation) on a graph paper with the help of matplotlib library
 with the following commands 

```python
plt.plot(x,y1, 'b') # x is the 2d line space we already defined. 'b' is for color 'blue'
plt.plot(x,y2, 'r')
plt.xlim([-2, 2])   # X axis is from -2 to 2
plt.ylim([-2, 2])   # y axis is from -2 to 2
plt.show()
```

# Transformation
we are transforming this circle with above given matrix as it it coefficients of x and y variables then we will plot the new point on the graphpaper

```python
def transformation(x,y):
    return 9*x + 4*y, 4*x +3*y
```

```python
x_new1, y_new1 = transformation(x,y1)
x_new2, y_new2 = transformation(x,y2)
```

```python
plt.plot(x_new1, y_new1, 'b')
plt.plot(x_new2, y_new2, 'r')
```

```python
eig_values, eig_vectors = np.linalg.eig(np.array([[9,4],[4,3]]))
print('Eigenvectors \n%s' %eig_vectors)
print('\nEigenvalues \n%s' %eig_values)
```

```python
soa = np.array([[0, 0,
                 eig_values[0] * eig_vectors[0][0],
                 eig_values[0] * eig_vectors[1][0]]])
soal = np.array([[0, 0,
                eig_values[1] * eig_vectors[0][1],
                eig_values[1] * eig_vectors[1][1]]])
X, Y, U, V = zip(*soa)
X1, Y1, U1, V1 = zip(*soal)
```

```python
plt.plot(x,y1, 'b') # x is the 2d line space we already defined. 'b' is for color 'blue'
plt.plot(x,y2, 'r')
plt.quiver(X,Y,U,V, angles = 'xy', scale_units='xy', scale = 1)
plt.quiver(X1,Y1,U1,V1, angles = 'xy', scale_units='xy', scale = 1)
plt.xlim([-1, 10])   # X axis is from -2 to 2
plt.ylim([-1, 10])   # y axis is from -2 to 2
plt.draw()
plt.show()
```

```python
plt.plot(x_new1,y_new1, 'b') # x is the 2d line space we already defined. 'b' is for color 'blue'
plt.plot(x_new2,y_new2, 'r')
plt.quiver(X,Y,U,V, angles = 'xy', scale_units='xy', scale = 1)
plt.quiver(X1,Y1,U1,V1, angles = 'xy', scale_units='xy', scale = 1)
#plt.xlim([-1, 10])   # X axis is from -1 to 10
#plt.ylim([-1, 10])   # y axis is from -1 to 10
plt.draw()
plt.show()
```
